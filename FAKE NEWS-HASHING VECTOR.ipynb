```python
import pandas as pd
import numpy as np
import itertools
#converting text data into numerical feature vectors.
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('news.csv', index_col=None)
```

```python
df
```

```python
dataset=df.drop("Unnamed: 0",axis=1)
```

```python
dataset
```

```python
x=dataset["text"]
y=dataset["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=53)
```

```python
hashing_vectorizer = HashingVectorizer(stop_words='english')
```

```python
hashing_vectorizer
```

```python
count_train = hashing_vectorizer.fit_transform(X_train)
```

```python
count_train
```

```python
print(count_train)
```

```python
count_test = hashing_vectorizer.transform(X_test)
```

```python
print(count_test)
```

```python
clf = MultinomialNB()

clf.fit(count_train, y_train) #MODEL CREATE
pred = clf.predict(count_test)
score = metrics.accuracy_score(y_test, pred) #EVALUATION METRIC
print("accuracy:   %0.3f" % score)

cm = metrics.confusion_matrix(y_test, pred, labels=['FAKE', 'REAL'])
```
