```python
 def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

#### TOPIC MODELLING

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')      # vector created  TEXT TO NUMBER
dtm=tfidf.fit_transform(df["Chat"])
```

```python
from sklearn.decomposition import NMF               #NON MATRIX FACTORIZATION ALGM
nmf_model=NMF(n_components=5,random_state=42)    #5 topics need
nmf_model.fit(dtm)
```

```python
# Retrieve the top words for each topic
for index, topic in enumerate(nmf_model.components_):
    # Use get_feature_names_out() to retrieve feature names
    results = [tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]]
    print(f"Topic {index + 1}: {', '.join(results)}")
```

```python
print("Shape of DTM:", dtm.shape)
print("Feature names:", tfidf.get_feature_names_out())
```

```python
topic_results=nmf_model.transform(dtm)
df["Topic"]=topic_results.argmax(axis=1)
```

#### Latent Dirichlet Allocation (LDA)

LDA is a popular method for discovering abstract topics in a collection of documents.
Gensim is a popular open-source library for topic modeling and document retrieval in Python. It provides efficient implementations of various topic modeling algorithms, including Latent Dirichlet Allocation (LDA) and Non-Negative Matrix Factorization (NMF).

Before applying topic modeling, you need to prepare your text corpus. Gensim provides utilities to preprocess the text, such as tokenization, stopword removal, and lemmatization. Here's an example:

```python
documents = df["Chat"]
```

```python
# Define stop words
stop_words = set(stopwords.words('english'))

# Tokenization and stop word removal
texts = [[word for word in word_tokenize(doc.lower()) if word.isalnum() and word not in stop_words] for doc in documents]
```

```python
import pandas as pd
import numpy as np
import gensim
from gensim import corpora
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize  #word_tokenize-splitting a string of text into individual words or tokens.
import nltk   #nltk provides word_tokenize

# Download NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Create a dictionary representation of the documents
dictionary = corpora.Dictionary(texts)

# Create a bag-of-words representation of the documents
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
from gensim.models import LdaModel

# Set parameters
num_topics = 2  # Number of topics
lda_model = LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=10, random_state=42)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")
```

#### enumerate() function is used to get both the index (i) and the document (doc) for each iteration.
#### Inside the loop, for each document, the code calls the get_document_topics() method of the lda_model.
#### This method takes a document (doc) as input and returns a list of tuples, where each tuple contains the topic ID and the proportion of the document belonging to that topic.
#### The resulting list of tuples is printed along with the document index (i).

#### Topic 0 emphasizes informal, possibly personal communication, indicating interactions that may involve scheduling or confirming plans.
Topic 1 focuses on more formal business communication, highlighting discussions about meetings and project details.


```python
# Assign topics to documents
for i, doc in enumerate(corpus):
    print(f"Document {i}: {lda_model.get_document_topics(doc)}")
```

#### Document 0 is assigned to both topic 0 and topic 1, with equal proportions (0.5 each).
Document 1 is predominantly assigned to topic 0 (0.8) and has a smaller proportion of topic 1 (0.2).
Document 2 is mainly assigned to topic 1 (0.7) and has a smaller proportion of topic 2 (0.3).

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF

# Define a diverse corpus
corpus = [
    "Machine learning is a fascinating field.",
    "Natural language processing enables computers to understand human language.",
    "Deep learning is a subset of machine learning that uses neural networks.",
    "Data science combines statistics and computer science to analyze data."
]

# Create and fit the TfidfVectorizer
tfidf = TfidfVectorizer(max_df=0.85, min_df=1, stop_words='english')
dtm = tfidf.fit_transform(corpus)  # Fit and transform the corpus

# Fit the NMF model to the DTM
n_components = 2  # Adjust the number of topics as needed
nmf_model = NMF(n_components=n_components, random_state=42)
nmf_model.fit(dtm)

# Transform the DTM to get topic distributions
topic_results = nmf_model.transform(dtm)

# Create a DataFrame to hold the results
df = pd.DataFrame(corpus, columns=['Document'])
df['Topic'] = topic_results.argmax(axis=1)  # Assign the topic with the highest score

# Display the DataFrame with assigned topics
print(df)
```
